```python
import os, random, imageio, shutil
from collections import deque

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from tqdm import trange

# -----------------------
# 1) ENVIRONMENT
# -----------------------

class InventoryEnv:
    def __init__(self, min_inv=-50, max_inv=50, T=50):
        self.min_inv = min_inv
        self.max_inv = max_inv
        self.T       = T
        self.reset()

    def reset(self):
        self.s = random.randint(self.min_inv, self.max_inv)
        self.t = 0
        return np.array([self.s, self.t], dtype=np.float32)

    def step(self, a):
        w = random.randint(0, 10)
        s2 = self.s + a - w
        s2 = max(self.min_inv, min(self.max_inv, s2))

        p, a_cost, b_cost = 1, 1, 2
        order_cost = p * a
        inv_cost   = a_cost * s2 if s2 >= 0 else b_cost * (-s2)
        cost       = order_cost + inv_cost
        r = cost

        self.t += 1
        done = (self.t >= self.T)
        self.s = s2
        obs = np.array([self.s, self.t if not done else 0], dtype=np.float32)
        return obs, r, done

# -----------------------
# 2) Q-NETWORK
# -----------------------

class DQN(nn.Module):
    def __init__(self, state_dim, num_actions):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(state_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, num_actions)
        )
    def forward(self, x):
        return self.net(x)

# -----------------------
# 3) REPLAY BUFFER
# -----------------------
class ReplayBuffer:
    def __init__(self, cap): self.buf = deque(maxlen=cap)
    def push(self, *args):    self.buf.append(tuple(args))
    def sample(self, k):
        batch = random.sample(self.buf, k)
        return zip(*batch)
    def __len__(self): return len(self.buf)

# -----------------------
# 4) HYPERPARAMETERS
# -----------------------
num_actions      = 10
state_dim        = 2
T                = 100
min_inv, max_inv = -50, 50
gamma            = 1.0
eps_start        = 0.9
eps_end          = 0.05
eps_decay        = 5000
buffer_cap       = 50_000
batch_size       = 32
lr               = 1e-3
target_update    = 10
episodes         = 3000

# -----------------------
# 5) SETUP
# -----------------------
device     = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
env        = InventoryEnv(min_inv, max_inv, T)
policy_net = DQN(state_dim, num_actions).to(device)
target_net = DQN(state_dim, num_actions).to(device)
target_net.load_state_dict(policy_net.state_dict())
opt        = optim.Adam(policy_net.parameters(), lr=lr)
buf        = ReplayBuffer(buffer_cap)

step_count = 0
eps        = eps_start

# For Q-value convergence check
q_convergence = []
prev_q_snapshot = None
check_convergence_interval = 50

# -----------------------
# 6) TRAIN
# -----------------------
for ep in trange(episodes, desc="Training"):
    obs = env.reset()
    done= False
    while not done:
        if random.random() < eps:
            a = random.randrange(num_actions)
        else:
            st = torch.FloatTensor(obs).unsqueeze(0).to(device)
            qv = policy_net(st)
            a  = int(qv.argmin(dim=1).item())

        obs2, r, done = env.step(a)
        buf.push(obs, a, r, obs2, done)
        obs = obs2

        if len(buf) >= batch_size:
            s_batch, a_batch, r_batch, s2_batch, d_batch = buf.sample(batch_size)
            s_v  = torch.FloatTensor(s_batch).to(device)
            a_v  = torch.LongTensor(a_batch).to(device)
            r_v  = torch.FloatTensor(r_batch).to(device)
            s2_v = torch.FloatTensor(s2_batch).to(device)
            d_v  = torch.FloatTensor(d_batch).to(device)

            q_vals = policy_net(s_v).gather(1, a_v.unsqueeze(-1)).squeeze(-1)

            with torch.no_grad():
                best_actions = policy_net(s2_v).argmin(dim=1)
                q2_selected = target_net(s2_v).gather(1, best_actions.unsqueeze(-1)).squeeze(-1)
                target = r_v + gamma * q2_selected * (1 - d_v)

            loss = nn.MSELoss()(q_vals, target)
            opt.zero_grad(); loss.backward(); opt.step()

            step_count += 1
            if step_count % target_update == 0:
                target_net.load_state_dict(policy_net.state_dict())

    eps = eps

    if ep % check_convergence_interval == 0:
        with torch.no_grad():
            q_snapshot = []
            for s in range(min_inv, max_inv+1):
                st = torch.FloatTensor([[s, 0]]).to(device)
                qs = policy_net(st).cpu().numpy()[0]
                q_snapshot.append(qs)
            q_snapshot = np.array(q_snapshot)

        if prev_q_snapshot is not None:
            diff = np.abs(prev_q_snapshot - q_snapshot)
            max_diff = np.max(diff)
            q_convergence.append(max_diff)
            print(f"[Episode {ep}] Max Q-value diff at t=0: {max_diff:.6f}")
        prev_q_snapshot = q_snapshot

print("Training complete")

# -----------------------
# Plot Q-value convergence
# -----------------------
if q_convergence:
    plt.figure(figsize=(6, 4))
    plt.plot(np.arange(len(q_convergence)) * check_convergence_interval, q_convergence, marker='o')
    plt.xlabel("Episode")
    plt.ylabel("Max ΔQ at t=0")
    plt.title("Q-value Convergence")
    plt.grid()
    plt.tight_layout()
    plt.show()

# -----------------------
# Extract Q-table & V(s,t) with fix for t=T-1
# -----------------------
policy_net.eval()
num_states = max_inv - min_inv + 1
Q_tab  = np.zeros((T, num_states, num_actions), dtype=np.float32)
V_tab  = np.zeros((T, num_states), dtype=np.float32)
pi_tab = np.zeros((T, num_states), dtype=int)

with torch.no_grad():
    for t in range(T):
        for i, s in enumerate(range(min_inv, max_inv+1)):
            if t == T - 1:
                Q_tab[t, i] = np.zeros(num_actions)
                V_tab[t, i] = 0
                pi_tab[t, i] = 0
            else:
                st = torch.FloatTensor([[s, t]]).to(device)
                qs = policy_net(st).cpu().numpy()[0]
                Q_tab[t, i] = qs
                a_opt = int(np.argmin(qs))
                pi_tab[t, i] = a_opt
                V_tab[t, i] = qs[a_opt]

# -----------------------
# 9) ROLLOUT DEMO
# -----------------------
print("\nTime | Inv | Demand | Action")
print("------------------------------------")
s = 0
for t in range(T):
    w = random.randint(0,10)
    a = pi_tab[t, s-min_inv]
    print(f"{t:4d} | {s:4d} | {w:6d} | {a:6d}")
    s = max(min_inv, min(max_inv, s + a - w))

# -----------------------
# 11) GIF of V(s) over t
# -----------------------
states = np.arange(min_inv, max_inv+1)
os.makedirs("frames", exist_ok=True)
for t in range(T):
    plt.figure(figsize=(4,3))
    plt.plot(states, V_tab[t], marker='o')
    plt.title(f"t={t}"); plt.tight_layout()
    fn = f"frames/{t:03d}.png"
    plt.savefig(fn); plt.close()

with imageio.get_writer("vf_evolution_ar_17.gif", mode="I", duration=0.1) as w:
    for t in range(T):
        w.append_data(imageio.imread(f"frames/{t:03d}.png"))

shutil.rmtree("frames")
print("GIF saved as vf_evolution_ar_17.gif")
```

```python
# Updated Double DQN for Inventory Control with Smooth V(s,t) Decay

import os, random, imageio, shutil
from collections import deque
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from tqdm import trange

# -----------------------
# 1) ENVIRONMENT
# -----------------------
class InventoryEnv:
    def __init__(self, min_inv=-50, max_inv=50, T=50):
        self.min_inv = min_inv
        self.max_inv = max_inv
        self.T = T
        self.reset()

    def reset(self):
        self.s = random.randint(self.min_inv, self.max_inv)
        self.t = 0
        return np.array([self.s, self.t], dtype=np.float32)

    def step(self, a):
        w = random.randint(0, 10)
        s2 = self.s + a - w
        s2 = max(self.min_inv, min(self.max_inv, s2))

        p, a_cost, b_cost = 1, 1, 2
        order_cost = p * a
        inv_cost = a_cost * s2 if s2 >= 0 else b_cost * (-s2)
        cost = order_cost + inv_cost
        r = cost

        self.t += 1
        done = (self.t >= self.T)
        self.s = s2
        obs = np.array([self.s, self.t], dtype=np.float32)
        return obs, r, done

# -----------------------
# 2) Q-NETWORK
# -----------------------
class DQN(nn.Module):
    def __init__(self, state_dim, num_actions):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(state_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, num_actions)
        )
    def forward(self, x):
        return self.net(x)

# -----------------------
# 3) REPLAY BUFFER
# -----------------------
class ReplayBuffer:
    def __init__(self, cap, n_step=1, gamma=1.0):
        self.buffer = deque(maxlen=cap)
        self.n_step = n_step
        self.gamma = gamma
        self.n_step_buffer = deque(maxlen=n_step)

    def _get_n_step_transition(self):
        r, s2, done = 0, None, False
        for idx, (_, _, r_i, s2_i, d_i) in enumerate(self.n_step_buffer):
            r += (self.gamma ** idx) * r_i
            s2, done = s2_i, d_i
            if d_i: break
        s, a = self.n_step_buffer[0][:2]
        return (s, a, r, s2, done)

    def push(self, s, a, r, s2, done):
        self.n_step_buffer.append((s, a, r, s2, done))
        if len(self.n_step_buffer) == self.n_step:
            self.buffer.append(self._get_n_step_transition())
        if done:
            while self.n_step_buffer:
                self.buffer.append(self._get_n_step_transition())
                self.n_step_buffer.popleft()

    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        return zip(*batch)

    def __len__(self):
        return len(self.buffer)


# -----------------------
# 4) HYPERPARAMETERS
# -----------------------
num_actions = 10
state_dim = 2
T = 50
min_inv, max_inv = -50, 50
gamma = 1.0
eps_start = 0.9
eps_end = 0.05
eps_decay = 5000
buffer_cap = 50_000
batch_size = 32
lr = 1e-3
target_update = 10
episodes = 2000
n_step = 3

# -----------------------
# 5) SETUP
# -----------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
env = InventoryEnv(min_inv, max_inv, T)
policy_net = DQN(state_dim, num_actions).to(device)
target_net = DQN(state_dim, num_actions).to(device)
target_net.load_state_dict(policy_net.state_dict())
opt = optim.Adam(policy_net.parameters(), lr=lr)
buf = ReplayBuffer(buffer_cap, n_step=n_step, gamma=gamma)

step_count = 0
eps = eps_start

q_convergence = []
prev_q_snapshot = None
check_convergence_interval = 50

# -----------------------
# 6) TRAIN
# -----------------------

for ep in trange(episodes, desc="Training"):
    obs = env.reset()
    done = False

    while not done:
        if random.random() < eps:
            a = random.randrange(num_actions)
        else:
            st = torch.FloatTensor(obs).unsqueeze(0).to(device)
            qv = policy_net(st)
            a = int(qv.argmin(dim=1).item())

        obs2, r, done = env.step(a)
        buf.push(obs, a, r, obs2, done)
        obs = obs2

        if len(buf) >= batch_size:
            s_batch, a_batch, r_batch, s2_batch, d_batch = buf.sample(batch_size)

            s_v = torch.FloatTensor(s_batch).to(device)
            a_v = torch.LongTensor(a_batch).to(device)
            r_v = torch.FloatTensor(r_batch).to(device)
            s2_v = torch.FloatTensor(s2_batch).to(device)
            d_v = torch.FloatTensor(d_batch).to(device)

            t_v = s_v[:, 1]  # extract time from state
            q_vals = policy_net(s_v).gather(1, a_v.unsqueeze(-1)).squeeze(-1)

            with torch.no_grad():
                best_actions = policy_net(s2_v).argmin(dim=1)
                q2_selected = target_net(s2_v).gather(1, best_actions.unsqueeze(-1)).squeeze(-1)
                non_terminal = (t_v < T - 1).float()  # only treat as terminal if time = T
                target = r_v + (gamma ** n_step) * q2_selected * non_terminal

            loss = nn.MSELoss()(q_vals, target)
            opt.zero_grad(); loss.backward(); opt.step()

            step_count += 1
            if step_count % target_update == 0:
                target_net.load_state_dict(policy_net.state_dict())

    if eps > eps_end:
        eps *= 0.9999

    if ep % check_convergence_interval == 0:
        with torch.no_grad():
            q_snapshot = []
            for s in range(min_inv, max_inv + 1):
                st = torch.FloatTensor([[s, T - 1]]).to(device)
                qs = policy_net(st).cpu().numpy()[0]
                q_snapshot.append(qs)
            q_snapshot = np.array(q_snapshot)

        if prev_q_snapshot is not None:
            diff = np.abs(prev_q_snapshot - q_snapshot)
            max_diff = np.max(diff)
            q_convergence.append(max_diff)
            print(f"[Episode {ep}] Max Q-value diff at t=T-1: {max_diff:.6f}")
        prev_q_snapshot = q_snapshot

print("Training complete")

# -----------------------
# Plot Q-value convergence
# -----------------------
if q_convergence:
    plt.figure(figsize=(6, 4))
    plt.plot(np.arange(len(q_convergence)) * check_convergence_interval, q_convergence, marker='o')
    plt.xlabel("Episode")
    plt.ylabel("Max ΔQ at t=T-1")
    plt.title("Q-value Convergence")
    plt.grid()
    plt.tight_layout()
    plt.show()

# -----------------------
# 7) EXTRACT Q-TABLE & V(s,t)
# -----------------------
policy_net.eval()
num_states = max_inv - min_inv + 1
Q_tab = np.zeros((T, num_states, num_actions), dtype=np.float32)
V_tab = np.zeros((T, num_states), dtype=np.float32)
pi_tab = np.zeros((T, num_states), dtype=int)

with torch.no_grad():
    for t in range(T):
        for i, s in enumerate(range(min_inv, max_inv + 1)):
            st = torch.FloatTensor([[s, t]]).to(device)
            qs = policy_net(st).cpu().numpy()[0]
            Q_tab[t, i] = qs
            a_opt = int(np.argmin(qs))
            pi_tab[t, i] = a_opt
            V_tab[t, i] = qs[a_opt]

# -----------------------
# 9) ROLLOUT DEMO
# -----------------------
print("\nTime | Inv | Demand | Action")
print("------------------------------------")
s = 0
for t in range(T):
    w = random.randint(0, 10)
    a = pi_tab[t, s - min_inv]
    print(f"{t:4d} | {s:4d} | {w:6d} | {a:6d}")
    s = max(min_inv, min(max_inv, s + a - w))

# -----------------------
# 11) GIF of V(s) over t
# -----------------------
states = np.arange(min_inv, max_inv + 1)
os.makedirs("frames", exist_ok=True)
for t in range(T):
    plt.figure(figsize=(4, 3))
    plt.plot(states, V_tab[t], marker='o')
    plt.title(f"t={t}"); plt.tight_layout()
    fn = f"frames/{t:03d}.png"
    plt.savefig(fn); plt.close()

with imageio.get_writer("vf_evolution_ar_118.gif", mode="I", duration=0.1) as w:
    for t in range(T):
        w.append_data(imageio.imread(f"frames/{t:03d}.png"))

shutil.rmtree("frames")
print("GIF saved as vf_evolution_ar_18.gif")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Redefine parameters (must match your training script)
min_inv, max_inv = -50, 50
T = 50
n_sims = 500
probe_states = list(range(min_inv, max_inv+1, 10))

# Compute bias
bias_records = []
for s0 in probe_states:
    cum_costs = []
    for _ in range(n_sims):
        env.s, env.t = s0, 0
        total = 0
        done = False
        while not done:
            a = pi_tab[env.t, env.s - min_inv]
            obs, c, done = env.step(a)
            total += c
        cum_costs.append(total)

    true_mean = np.mean(cum_costs)
    v_est     = V_tab[0, s0 - min_inv]
    bias      = v_est - true_mean
    bias_records.append({'s0': s0, 'Bias': bias})

df_bias = pd.DataFrame(bias_records)

# Plot bias vs initial state
plt.figure()
plt.plot(df_bias['s0'], df_bias['Bias'], marker='o')
plt.xlabel("Initial Inventory Level (s₀)")
plt.ylabel("Overestimation Bias")
plt.title("Double DQN Overestimation Bias vs. Initial State")
plt.grid()
plt.tight_layout()
plt.show()

# -----------------------
# 8) EVALUATE VALUE VS MC RETURN (n-step)
# -----------------------

def estimate_mc_return(pi_tab, num_rollouts=100):
    mc_vals = np.zeros((num_states,))
    mc_std = np.zeros((num_states,))
    for idx, s0 in enumerate(range(min_inv, max_inv + 1)):
        returns = []
        for _ in range(num_rollouts):
            s = s0
            total_cost = 0
            for t in range(T):
                a = pi_tab[t, s - min_inv]
                w = random.randint(0, 10)
                s2 = max(min_inv, min(max_inv, s + a - w))
                p, a_cost, b_cost = 1, 1, 2
                order_cost = p * a
                inv_cost = a_cost * s2 if s2 >= 0 else b_cost * (-s2)
                cost = order_cost + inv_cost
                total_cost += (gamma ** t) * cost
                s = s2
            returns.append(total_cost)
        mc_vals[idx] = np.mean(returns)
        mc_std[idx] = np.std(returns)
    return mc_vals, mc_std

V_est = V_tab[0]  # V(s, t=0) from Double DQN
V_mc, V_std = estimate_mc_return(pi_tab)

# -----------------------
# 10) PLOT VALUE VS MC RETURN
# -----------------------
plt.figure(figsize=(7, 5))
inv_states = np.arange(min_inv, max_inv + 1)
plt.errorbar(inv_states, V_mc, yerr=V_std, fmt='-o', label='Monte Carlo return')
plt.plot(inv_states, V_est, '-s', label='Estimated V(s₀) (Double DQN)')
plt.title(r"Estimated vs Actual $V(s_0, t=0)$")
plt.xlabel("Inventory Level (s₀)")
plt.ylabel("Value / Return (Cost)")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Updated Double DQN for Inventory Control with Smooth V(s,t) Decay

import os, random, imageio, shutil
from collections import deque
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from tqdm import trange

# -----------------------
# 1) ENVIRONMENT
# -----------------------
class InventoryEnv:
    def __init__(self, min_inv=-50, max_inv=50, T=50):
        self.min_inv = min_inv
        self.max_inv = max_inv
        self.T = T
        self.reset()

    def reset(self):
        self.s = random.randint(self.min_inv, self.max_inv)
        self.t = 0
        return np.array([self.s, self.t], dtype=np.float32)

    def step(self, a):
        w = random.randint(0, 10)
        s2 = self.s + a - w
        s2 = max(self.min_inv, min(self.max_inv, s2))

        p, a_cost, b_cost = 1, 1, 2
        order_cost = p * a
        inv_cost = a_cost * s2 if s2 >= 0 else b_cost * (-s2)
        cost = order_cost + inv_cost
        r = cost

        self.t += 1
        done = (self.t >= self.T)
        self.s = s2
        obs = np.array([self.s, self.t], dtype=np.float32)
        return obs, r, done

# -----------------------
# 2) Q-NETWORK
# -----------------------
class DQN(nn.Module):
    def __init__(self, state_dim, num_actions):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(state_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, num_actions)
        )
    def forward(self, x):
        return self.net(x)

# -----------------------
# 3) REPLAY BUFFER
# -----------------------
class ReplayBuffer:
    def __init__(self, cap, n_step=1, gamma=1.0):
        self.buffer = deque(maxlen=cap)
        self.n_step = n_step
        self.gamma = gamma
        self.n_step_buffer = deque(maxlen=n_step)

    def _get_n_step_transition(self):
        r, s2, done = 0, None, False
        for idx, (_, _, r_i, s2_i, d_i) in enumerate(self.n_step_buffer):
            r += (self.gamma ** idx) * r_i
            s2, done = s2_i, d_i
            if d_i: break
        s, a = self.n_step_buffer[0][:2]
        return (s, a, r, s2, done)

    def push(self, s, a, r, s2, done):
        self.n_step_buffer.append((s, a, r, s2, done))
        if len(self.n_step_buffer) == self.n_step:
            self.buffer.append(self._get_n_step_transition())
        if done:
            while self.n_step_buffer:
                self.buffer.append(self._get_n_step_transition())
                self.n_step_buffer.popleft()

    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        return zip(*batch)

    def __len__(self):
        return len(self.buffer)


# -----------------------
# 4) HYPERPARAMETERS
# -----------------------
num_actions = 10
state_dim = 2
T = 50
min_inv, max_inv = -50, 50
gamma = 1.0
eps_start = 0.9
eps_end = 0.05
eps_decay = 5000
buffer_cap = 50_000
batch_size = 32
lr = 1e-3
target_update = 10
episodes = 2000
n_step = 3

# -----------------------
# 5) SETUP
# -----------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
env = InventoryEnv(min_inv, max_inv, T)
policy_net = DQN(state_dim, num_actions).to(device)
target_net = DQN(state_dim, num_actions).to(device)
target_net.load_state_dict(policy_net.state_dict())
opt = optim.Adam(policy_net.parameters(), lr=lr)
buf = ReplayBuffer(buffer_cap, n_step=n_step, gamma=gamma)

step_count = 0
eps = eps_start

q_convergence = []
prev_q_snapshot = None
check_convergence_interval = 50

# -----------------------
# 6) TRAIN
# -----------------------
 # -----------------------
# 6) TRAIN (Blended λ target)
# -----------------------

def compute_blended_target(policy_net, target_net, next_states, rewards, dones, time_steps,
                           nstep_returns, gamma, T, lambda_blend):
    with torch.no_grad():
        best_actions = policy_net(next_states).argmin(dim=1, keepdim=True)
        q_values = target_net(next_states).gather(1, best_actions).squeeze(-1)
        non_terminal = (time_steps < T - 1).float()
        one_step_target = rewards + gamma * q_values * non_terminal
        blended_target = lambda_blend * nstep_returns + (1 - lambda_blend) * one_step_target
        return blended_target

def compute_nstep_return(buf, gamma, n, idxs):
    n_returns = []
    for idx in idxs:
        R = 0
        discount = 1.0
        for n_idx in range(n):
            if idx + n_idx >= len(buf.buffer):
                break
            _, _, r, _, d = buf.buffer[idx + n_idx]
            R += discount * r
            discount *= gamma
            if d:
                break
        n_returns.append(R)
    return torch.FloatTensor(n_returns).to(device)


# Blending parameter
lambda_blend = 0.7
n_step = 3

for ep in trange(episodes, desc="Training"):
    obs = env.reset()
    done = False

    while not done:
        if random.random() < eps:
            a = random.randrange(num_actions)
        else:
            st = torch.FloatTensor(obs).unsqueeze(0).to(device)
            qv = policy_net(st)
            a = int(qv.argmin(dim=1).item())

        obs2, r, done = env.step(a)
        buf.push(obs, a, r, obs2, done)
        obs = obs2

        if len(buf) >= batch_size + n_step:
            # Sample indexes to compute n-step returns
            idxs = random.sample(range(len(buf) - n_step), batch_size)
            batch = [buf.buffer[idx] for idx in idxs]
            s_batch, a_batch, r_batch, s2_batch, d_batch = zip(*batch)

            s_v = torch.FloatTensor(s_batch).to(device)
            a_v = torch.LongTensor(a_batch).to(device)
            r_v = torch.FloatTensor(r_batch).to(device)
            s2_v = torch.FloatTensor(s2_batch).to(device)
            d_v = torch.FloatTensor(d_batch).to(device)
            t_v = s_v[:, 1]

            # Compute blended target using n-step returns
            nstep_returns = compute_nstep_return(buf, gamma, n_step, idxs)
            target = compute_blended_target(policy_net, target_net, s2_v, r_v, d_v, t_v,
                                            nstep_returns, gamma, T, lambda_blend)

            q_vals = policy_net(s_v).gather(1, a_v.unsqueeze(-1)).squeeze(-1)
            loss = nn.MSELoss()(q_vals, target)

            opt.zero_grad(); loss.backward(); opt.step()

            step_count += 1
            if step_count % target_update == 0:
                target_net.load_state_dict(policy_net.state_dict())

    if eps > eps_end:
        eps *= 0.9999

    if ep % check_convergence_interval == 0:
        with torch.no_grad():
            q_snapshot = []
            for s in range(min_inv, max_inv + 1):
                st = torch.FloatTensor([[s, T - 1]]).to(device)
                qs = policy_net(st).cpu().numpy()[0]
                q_snapshot.append(qs)
            q_snapshot = np.array(q_snapshot)

        if prev_q_snapshot is not None:
            diff = np.abs(prev_q_snapshot - q_snapshot)
            max_diff = np.max(diff)
            q_convergence.append(max_diff)
            print(f"[Episode {ep}] Max Q-value diff at t=T-1: {max_diff:.6f}")
        prev_q_snapshot = q_snapshot

print("Training complete")


# -----------------------
# Plot Q-value convergence
# -----------------------
if q_convergence:
    plt.figure(figsize=(6, 4))
    plt.plot(np.arange(len(q_convergence)) * check_convergence_interval, q_convergence, marker='o')
    plt.xlabel("Episode")
    plt.ylabel("Max ΔQ at t=T-1")
    plt.title("Q-value Convergence")
    plt.grid()
    plt.tight_layout()
    plt.show()

# -----------------------
# 7) EXTRACT Q-TABLE & V(s,t)
# -----------------------
policy_net.eval()
num_states = max_inv - min_inv + 1
Q_tab = np.zeros((T, num_states, num_actions), dtype=np.float32)
V_tab = np.zeros((T, num_states), dtype=np.float32)
pi_tab = np.zeros((T, num_states), dtype=int)

with torch.no_grad():
    for t in range(T):
        for i, s in enumerate(range(min_inv, max_inv + 1)):
            st = torch.FloatTensor([[s, t]]).to(device)
            qs = policy_net(st).cpu().numpy()[0]
            Q_tab[t, i] = qs
            a_opt = int(np.argmin(qs))
            pi_tab[t, i] = a_opt
            V_tab[t, i] = qs[a_opt]

# -----------------------
# 9) ROLLOUT DEMO
# -----------------------
print("\nTime | Inv | Demand | Action")
print("------------------------------------")
s = 0
for t in range(T):
    w = random.randint(0, 10)
    a = pi_tab[t, s - min_inv]
    print(f"{t:4d} | {s:4d} | {w:6d} | {a:6d}")
    s = max(min_inv, min(max_inv, s + a - w))

# -----------------------
# 11) GIF of V(s) over t
# -----------------------
states = np.arange(min_inv, max_inv + 1)
os.makedirs("frames", exist_ok=True)
for t in range(T):
    plt.figure(figsize=(4, 3))
    plt.plot(states, V_tab[t], marker='o')
    plt.title(f"t={t}"); plt.tight_layout()
    fn = f"frames/{t:03d}.png"
    plt.savefig(fn); plt.close()

with imageio.get_writer("vf_evolution_ar_118.gif", mode="I", duration=0.1) as w:
    for t in range(T):
        w.append_data(imageio.imread(f"frames/{t:03d}.png"))

shutil.rmtree("frames")
print("GIF saved as vf_evolution_ar_18.gif")

# -----------------------
# 8) EVALUATE VALUE VS MC RETURN (n-step)
# -----------------------

def evaluate_value_estimates(policy_net, env, pi_tab, min_inv, max_inv, T, num_rollouts=100):
    states = np.arange(min_inv, max_inv + 1)
    est_vals = []
    mc_vals = []
    mc_errs = []

    for s in states:
        # Estimated V(s, t=0)
        st = torch.FloatTensor([[s, 0]]).to(device)
        v_est = policy_net(st).cpu().detach().numpy()[0]
        v_s = v_est[pi_tab[0, s - min_inv]]
        est_vals.append(v_s)

        # Monte Carlo estimates
        returns = []
        for _ in range(num_rollouts):
            env.s, env.t = s, 0
            obs = np.array([s, 0], dtype=np.float32)
            total_cost = 0
            t = 0
            done = False
            while not done:
                a = pi_tab[t, env.s - min_inv]
                obs, r, done = env.step(a)
                total_cost += r
                t += 1
            returns.append(total_cost)

        mc_mean = np.mean(returns)
        mc_std = np.std(returns)
        mc_vals.append(mc_mean)
        mc_errs.append(mc_std)

    return states, est_vals, mc_vals, mc_errs

states, est_vals, mc_vals, mc_errs = evaluate_value_estimates(policy_net, InventoryEnv(min_inv, max_inv, T), pi_tab, min_inv, max_inv, T)

# -----------------------
# Plot Estimated vs Monte Carlo Value
# -----------------------
plt.figure(figsize=(8, 5))
plt.errorbar(states, mc_vals, yerr=mc_errs, label="Monte Carlo return", fmt='-o', capsize=4)
plt.plot(states, est_vals, label="Estimated $V(s_0)$ (Double DQN)", marker='s')
plt.xlabel("Initial Inventory $s_0$")
plt.ylabel("Value / Return (Cost)")
plt.title("Estimated vs Actual $V(s_0, t = 0)$")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Integrated Double DQN vs DQN Bias Comparison for Inventory Control

import os, random, imageio, shutil
from collections import deque
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from tqdm import trange
import pandas as pd

# -----------------------
# 1) ENVIRONMENT
# -----------------------
class InventoryEnv:
    def __init__(self, min_inv=-50, max_inv=50, T=50):
        self.min_inv = min_inv
        self.max_inv = max_inv
        self.T = T
        self.reset()

    def reset(self):
        self.s = random.randint(self.min_inv, self.max_inv)
        self.t = 0
        return np.array([self.s, self.t], dtype=np.float32)

    def step(self, a):
        w = random.randint(0, 10)
        s2 = self.s + a - w
        s2 = max(self.min_inv, min(self.max_inv, s2))

        p, a_cost, b_cost = 1, 1, 2
        order_cost = p * a
        inv_cost = a_cost * s2 if s2 >= 0 else b_cost * (-s2)
        cost = order_cost + inv_cost
        r = cost

        self.t += 1
        done = (self.t >= self.T)
        self.s = s2
        obs = np.array([self.s, self.t], dtype=np.float32)
        return obs, r, done

# -----------------------
# 2) Q-NETWORK
# -----------------------
class DQN(nn.Module):
    def __init__(self, state_dim, num_actions):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(state_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, num_actions)
        )
    def forward(self, x):
        return self.net(x)

# -----------------------
# 3) REPLAY BUFFER
# -----------------------
class ReplayBuffer:
    def __init__(self, cap): self.buf = deque(maxlen=cap)
    def push(self, *args): self.buf.append(tuple(args))
    def sample(self, k):
        batch = random.sample(self.buf, k)
        return zip(*batch)
    def __len__(self): return len(self.buf)

# -----------------------
# 4) HYPERPARAMETERS
# -----------------------
num_actions = 10
state_dim = 2
T = 50
min_inv, max_inv = -50, 50
gamma = 1.0
eps_start = 0.9
eps_end = 0.05
eps_decay = 5000
buffer_cap = 50_000
batch_size = 32
lr = 1e-3
target_update = 10
episodes = 2000

# -----------------------
# 5) SETUP
# -----------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
env = InventoryEnv(min_inv, max_inv, T)

# Double DQN
policy_net = DQN(state_dim, num_actions).to(device)
target_net = DQN(state_dim, num_actions).to(device)
target_net.load_state_dict(policy_net.state_dict())
opt = optim.Adam(policy_net.parameters(), lr=lr)

# Standard DQN
dqn_policy_net = DQN(state_dim, num_actions).to(device)
dqn_target_net = DQN(state_dim, num_actions).to(device)
dqn_target_net.load_state_dict(dqn_policy_net.state_dict())
dqn_opt = optim.Adam(dqn_policy_net.parameters(), lr=lr)

buf = ReplayBuffer(buffer_cap)
step_count = 0
eps = eps_start

# -----------------------
# 6) TRAINING LOOP
# -----------------------
for ep in trange(episodes, desc="Training"):
    obs = env.reset()
    done = False
    while not done:
        if random.random() < eps:
            a = random.randrange(num_actions)
        else:
            st = torch.FloatTensor(obs).unsqueeze(0).to(device)
            qv = policy_net(st)
            a = int(qv.argmin(dim=1).item())

        obs2, r, done = env.step(a)
        buf.push(obs, a, r, obs2, done)
        obs = obs2

        if len(buf) >= batch_size:
            s_batch, a_batch, r_batch, s2_batch, d_batch = buf.sample(batch_size)

            s_v = torch.FloatTensor(s_batch).to(device)
            a_v = torch.LongTensor(a_batch).to(device)
            r_v = torch.FloatTensor(r_batch).to(device)
            s2_v = torch.FloatTensor(s2_batch).to(device)
            d_v = torch.FloatTensor(d_batch).to(device)

            t_v = s_v[:, 1]

            # Double DQN update
            q_vals = policy_net(s_v).gather(1, a_v.unsqueeze(-1)).squeeze(-1)
            with torch.no_grad():
                best_actions = policy_net(s2_v).argmin(dim=1)
                q2_selected = target_net(s2_v).gather(1, best_actions.unsqueeze(-1)).squeeze(-1)
                non_terminal = (t_v < T - 1).float()
                target = r_v + gamma * q2_selected * non_terminal
            loss = nn.MSELoss()(q_vals, target)
            opt.zero_grad(); loss.backward(); opt.step()

            # Standard DQN update
            q_vals_dqn = dqn_policy_net(s_v).gather(1, a_v.unsqueeze(-1)).squeeze(-1)
            with torch.no_grad():
                q2_dqn = dqn_target_net(s2_v).min(dim=1)[0]
                target_dqn = r_v + gamma * q2_dqn * non_terminal
            loss_dqn = nn.MSELoss()(q_vals_dqn, target_dqn)
            dqn_opt.zero_grad(); loss_dqn.backward(); dqn_opt.step()

            step_count += 1
            if step_count % target_update == 0:
                target_net.load_state_dict(policy_net.state_dict())
                dqn_target_net.load_state_dict(dqn_policy_net.state_dict())

    if eps > eps_end:
        eps *= 0.9999

# -----------------------
# 7) EXTRACT VALUE TABLES
# -----------------------
policy_net.eval()
dqn_policy_net.eval()
num_states = max_inv - min_inv + 1
V_double = np.zeros((T, num_states))
V_dqn = np.zeros((T, num_states))
pi_double = np.zeros((T, num_states), dtype=int)
pi_dqn = np.zeros((T, num_states), dtype=int)

with torch.no_grad():
    for t in range(T):
        for i, s in enumerate(range(min_inv, max_inv + 1)):
            st = torch.FloatTensor([[s, t]]).to(device)
            q_ddqn = policy_net(st)[0].cpu().numpy()
            q_dqn = dqn_policy_net(st)[0].cpu().numpy()

            pi_double[t, i] = np.argmin(q_ddqn)
            pi_dqn[t, i] = np.argmin(q_dqn)
            V_double[t, i] = q_ddqn[pi_double[t, i]]
            V_dqn[t, i] = q_dqn[pi_dqn[t, i]]

# -----------------------
# 8) BIAS ANALYSIS
# -----------------------
n_sims = 500
probe_states = list(range(min_inv, max_inv+1, 10))
bias_records = []

for s0 in probe_states:
    cum_ddqn = []
    cum_dqn = []
    for _ in range(n_sims):
        env.s, env.t = s0, 0
        total_ddqn = 0
        total_dqn = 0
        done = False
        s = env.s
        t = env.t
        while not done:
            a_ddqn = pi_double[t, s - min_inv]
            a_dqn = pi_dqn[t, s - min_inv]
            obs_ddqn, r_ddqn, done = env.step(a_ddqn)
            total_ddqn += r_ddqn
        cum_ddqn.append(total_ddqn)

        env.s, env.t = s0, 0
        done = False
        s = env.s
        t = env.t
        while not done:
            a_dqn = pi_dqn[t, s - min_inv]
            obs_dqn, r_dqn, done = env.step(a_dqn)
            total_dqn += r_dqn
        cum_dqn.append(total_dqn)

    true_mc = np.mean(cum_ddqn)  # Same MC reference for both
    bias_records.append({
        's0': s0,
        'Bias_DoubleDQN': V_double[0, s0 - min_inv] - true_mc,
        'Bias_DQN': V_dqn[0, s0 - min_inv] - true_mc
    })

df_bias = pd.DataFrame(bias_records)

# -----------------------
# 9) PLOT BIASES
# -----------------------
plt.figure(figsize=(8, 5))
plt.plot(df_bias['s0'], df_bias['Bias_DoubleDQN'], 'o-', label='Double DQN')
plt.plot(df_bias['s0'], df_bias['Bias_DQN'], 's--', label='Standard DQN')
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.xlabel("Initial Inventory Level $s_0$")
plt.ylabel("Overestimation Bias")
plt.title("Bias Comparison: Double DQN vs DQN")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
